```python
#Scale the data via MinMax Scaler
# reshape (-1,1 ) means change the array df to single column array with dimensions such that number of elements in orignal array remains the same. eg an array 2,4 reshaped to(-1,1) will change the array to (8,1).
import numpy as np
from sklearn.preprocessing import MinMaxScaler
scaler = MinMaxScaler(feature_range=(0,1))
df2 = scaler.fit_transform(np.array(df1).reshape(-1,1))
df2
```

```python
train_size=int(len(df2)*0.65)
test_size=len(df2)-train_size
train_data = df2[0:train_size,:]
test_data = df2[train_size:len(df2),:]

train_size,test_size
len(train_data)
```

```python
import numpy
def create_dataset(dataset,time_steps=1):
    data_X,data_Y = [],[]
    for i in range(len(dataset)-time_steps-1):
        
        a = dataset[i:(i+time_steps),0]
        data_X.append(a)
        data_Y.append(dataset[(i+time_steps),0])
    
    return numpy.array(data_X), numpy.array(data_Y)
    
```

```python
# reshape into X=t,t+1,t+2,t+3 and Y=t+4
time_steps = 100
X_train, y_train = create_dataset(train_data, time_steps)
X_test, y_test = create_dataset(test_data, time_steps)

X_train.shape
```

```python
X_train.shape
```

```python
#Changing to 3 dimensions for LSTM ,  Check Why ?

X_train = X_train.reshape(X_train.shape[0],X_train.shape[1],1)
X_test = X_test.reshape(X_test.shape[0],X_test.shape[1],1)
                    
```

```python
# Create Stacked LSTM
from  tensorflow.keras.models import Sequential
from  tensorflow.keras.layers import Dense,LSTM
```

```python
model = Sequential()

model.add(LSTM(50,return_sequences =True,input_shape = (100,1)))
model.add(LSTM(50,return_sequences =True))
model.add(LSTM(50))
model.add(Dense(1))

model.compile(loss = 'mean_squared_error',optimizer = 'adam',)

```

model.fit(X_train,y_train,validation_data = (X_test,y_test),epochs =100, batch_size =64,verbose =1)
