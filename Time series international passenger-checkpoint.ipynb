# Stationarity check 

```python
def test_stationarity(time_series):
    result = adfuller(time_series)
    labels  = ['ADF_Test_statistic', 'p-value','Lags used','Number ofobservations' ]
    for labels,values in zip(labels,result):
        print(labels+' : '+str(values))
    if result[1] < 0.05:
        print("Time series is stationary")
    else: 
        print("Make it stationary,Dude !! ")
    
    
test_stationarity(df['#Passengers'])
```

```python
rolling_std = df.rolling(window=12).std()
plt.plot(df)
plt.plot(rolling_mean,color = 'yellow')
plt.plot(rolling_std,color = 'green')
```

```python
df_log  = np.log(df)
plt.plot(df_log)
```

```python
Moving_average = df_log.rolling(window=12).mean()
plt.plot(df_log)
plt.plot(Moving_average,color = 'Blue')

```

```python
ts = df_log - Moving_average
ts.dropna(inplace = True)
ts.head()
plt.plot(ts)
test_stationarity(ts)
```

```python
exp_data=df_log.ewm(halflife=12, min_periods=0, adjust=True).mean()
plt.plot(df_log)
plt.plot(exp_data, color='black')
```

```python
exp_data_diff = df_log - exp_data 
plt.plot(df_log)
test_stationarity(exp_data_diff)
```

```python
# to remove seasonality , need to do differencing

df_log_shift = df_log - df_log.shift(12)
df_log_shift.dropna(inplace=True)
plt.plot(df_log_shift)

test_stationarity(df_log_shift)
```

```python
# to remove seasonality 
from statsmodels.tsa.seasonal import seasonal_decompose

decomp_ts = seasonal_decompose(df_log)

trend = decomp_ts.trend
seasonal = decomp_ts.seasonal
residual = decomp_ts.resid

plt.subplot(411)
plt.plot(df_log, label='Original')
plt.legend(loc='best')
plt.subplot(412)
plt.plot(trend, label='Trend')
plt.legend(loc='best')
plt.subplot(413)
plt.plot(seasonal,label='Seasonality')
plt.legend(loc='best')
plt.subplot(414)
plt.plot(residual, label='Residuals')
plt.legend(loc='best')
plt.tight_layout()

```

```python
ts_decomp_resid = residual
test_stationarity(ts_decomp_resid)

plt.figure( figsize=(20,5))
plt.plot(ts_decomp_resid)

ts_decomp_resid.dropna(inplace=True)
ts_decomp_resid.head()
```

```python
acf_graph = acf(ts_decomp_resid.iloc[0:],nlags =30)
pacf_graph = pacf(ts_decomp_resid.iloc[0:],nlags =30,method='ols')

plt.figure(figsize=(20,10))
plt.subplot(121)
plt.plot(acf_graph)
plt.axhline(y=0,linestyle='--',color='green')
plt.axhline(y=-1.96/np.sqrt(len(ts_decomp_resid)),linestyle='--',color='green')
plt.axhline(y=1.96/np.sqrt(len(ts_decomp_resid)),linestyle='--',color='green')
plt.title('Autocorrelation Function')

plt.subplot(122)
plt.plot(pacf_graph)
plt.axhline(y=0,linestyle='--',color='green')
plt.axhline(y=-1.96/np.sqrt(len(ts_decomp_resid)),linestyle='--',color='green')
plt.axhline(y=1.96/np.sqrt(len(ts_decomp_resid)),linestyle='--',color='green')
plt.title('Partial Autocorrelation Function')
```
